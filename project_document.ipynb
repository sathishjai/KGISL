## introduction

This project involves an exploratory data analysis (EDA) of laptop characteristics and their pricing. The dataset comprises information on various laptop attributes such as brand, processor, RAM, storage, screen size, and price. The objective is to uncover insights into how different features influence laptop prices, identify trends, and provide data-driven recommendations.

## Aim

The aim of this project is to analyze the relationship between laptop features and their prices to identify key factors that significantly influence pricing. This analysis will help consumers make informed decisions and assist retailers in optimizing their product offerings and pricing strategies.

## Business Problem / Problem Statement

Laptops are essential for various users, from students to professionals, but choosing the right laptop at the right price can be challenging due to the vast array of options. This project aims to address the business problem of understanding the key determinants of laptop pricing. By analyzing the dataset, we aim to identify which features most significantly impact price, helping retailers to tailor their inventory and pricing strategies and consumers to make cost-effective choices.

```python
def remove_outliers_iqr(df2, feature):
    Q1=df2[feature].quantile(0.25)
    Q3=df2[feature].quantile(0.75)
    IQR=Q3-Q1
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR
    return df2 [(df2 [feature] >= lower_bound) & (df2[feature] <= upper_bound)]
```

## Overall Insights Obtained from Analysis



Key Factors Influencing Price:

    Processor type, RAM, and screen size are the primary determinants of laptop pricing. High-end processors and larger RAM capacities lead to higher prices.
    Brand significantly affects pricing, with premium brands like Apple charging higher prices for similar specifications compared to other brands.

Market Trends:

    The majority of laptops are priced under $1,500, with a significant portion in the $500-$1,000 range, reflecting a demand for mid-range laptops.
    Brands like Acer and Asus offer competitive pricing for budget-conscious consumers, whereas premium brands dominate the higher price segment.

Consumer Insights:

    Laptops with a balance of performance and price, such as those with moderate RAM and mid-range processors, offer good value for money.
    Consumers looking for premium features should expect to pay more for high-end processors and larger screen sizes.

Recommendations for Manufacturers:

    Focus on offering a range of laptops across different price segments to cater to diverse consumer needs.
    Emphasize competitive pricing for mid-range laptops while highlighting the premium features in high-end models to justify higher prices.

## Conclusion

This comprehensive analysis revealed that processor type, RAM, and screen size are key drivers of laptop pricing. The insights gained provide valuable information for both consumers and manufacturers, guiding informed purchasing decisions and strategic pricing.
